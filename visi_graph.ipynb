```python
# pip3 install python-igraph
# pip3 install cairocffi
from igraph import *
import pandas as pd
from visibility_graph import visibility_graph
```

```python
acoes = pd.read_csv("gol.csv", decimal=".", header=None)
acoes_array = acoes.values[1:,4:5].astype(float)
acoes_df = pd.DataFrame(acoes_array)
td = acoes_df.head()

td = [i for i in td[0]]
td
```

```python
gg = Graph()

g = visibility_graph(td)
# O total como Ã© passado como argumento para a quantidade de vertices
gg.add_vertices(len(g.edges()))
gg.add_edges(g.edges())
```

```python
plot(gg)
plot(gg, vertex_label=td, vertex_color="blue") 
```
